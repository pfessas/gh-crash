```python
import pandas as pd
import numpy as np
from datetime import datetime
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle("df_2.pkl")
df = df.drop_duplicates()
```

### Plot

```python
# Function
def category_bloxpot(df,column):

    column_unique_values = (df[column].value_counts().index.tolist())
    column_unique_values.sort()
    
    temp_data = df[[column,'total_delay']]
    title    = column +': '+' - Box Plot of Total Delay Distribution'
    fig, ax = plt.subplots(figsize=(25,10))
          
    sns.boxplot(data=temp_data, x=column, y="total_delay", hue=column, showfliers = False,dodge =False,ax=ax)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 0.5),fontsize=15)
    plt.title(title, fontdict={'fontsize': 18})
    plt.ylabel('Seconds')
    plt.xlabel(column)
    plt.show()
```

```python
column='week_day_nbr'
category_bloxpot(df,column)
```

```python
column='quarter'
category_bloxpot(df,column)
```

```python
column='weekday_weekend'
category_bloxpot(df,column)
```
